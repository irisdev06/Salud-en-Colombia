# Limpiando los Datos üßπ
Decid√≠ no incluir las columnas:  
**"GHO (URL)" =** Es el url del c√≥digo del indicador y no afecta en nada su eliminaci√≥n.
**"DIMENSION (CODE)", "DIMENSION (NAME)", "Low", "High" =** Columnas relacionadas entre s√≠ que tienen el 91% de los datos nulos. 
Y elimin√© el index 0, ya que era texto de gu√≠a. 

```python
INDEX = PRESA01.drop(0)
SA01 = INDEX.drop(columns=["GHO (URL)", "DIMENSION (TYPE)",  "DIMENSION (CODE)", "DIMENSION (NAME)", "Low", "High"])
SA01.head(20)
```

# Analisis 
N√∫mero de camas por a√±o, para analizar c√≥mo ha incrementado. 

```python
a√±o_camas = SA01.groupby("STARTYEAR")["Numeric"].sum().sort_values()
a√±o_camas
plt.figure(figsize=(20, 6))
plt.scatter(a√±o_camas.index, a√±o_camas.values, c="purple")
plt.xlabel('Camas')
plt.ylabel('A√±o')
plt.title('A√±o vs Camas')
plt.show()
```
